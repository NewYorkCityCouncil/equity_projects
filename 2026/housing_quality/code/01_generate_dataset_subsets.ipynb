```python
import pandas as pd
```

# Occupied dataset

```python
occupied_dataset = pd.read_csv("../data/raw/occupied_puf_23.csv")
occupied_dataset
```

```python
housing_quality_columns = [
    "NOHEAT", # Indicates whether there was any heating breakdown in the previous winter, even if it was for less than 6 hours. 
    "NOHEAT_NUM", # Indicates the number of times the unit has a heating breakdown for 6 or more hours in the previous winter.
    "NOHOTWATER", # Indicates whether there was no hot water for at least 6 hours anytime in the previous year. 
    "LEAKS", # Indicates whether there were any leaks from outside the unit in the past year
    "MOLD", # Indicates whether there has been any mold in the unit in the previous year. 
    "MUSTY", # Indicates the frequency that the occupant smelled musty or moldy smells in the unit in the previous year. 
    "ELEVATOR_BROK", # Indicates whether any elevator in the building was not working for 6 or more hours in the previous year
    "ELEVATOR_ALLBROK", # Indicates whether all elevators in the building were not working for 6 or more hours in the previous year. 
    "RODENTS_UNIT", # Indicates if there was any evidence of mice or rats in the unit in the past three months. 
    "RODENTS_BUILD", # Indicates if there was any evidence of mice or rats in the building where the unit is located in the past three months. 
    "TOILET_BROK", # Indicates that none of the toilets in the unit were working for 6 hours or more in the previous three months. 
    "ROACHES_NUM", # Indicates the number of roaches the occupant(s) saw on a typical day in the last month. 
    "WALLHOLES", # Indicates if there were holes in the walls or ceiling inside the unit at the time of the NYCHVS interview. 
    "FLOORHOLES", # Indicates if there were holes in the floors inside the unit at the time of the NYCHVS interview.
    "PEELPAINT", # Indicates if there was any broken plaster or peeling paint inside the unit at the time of the NYCHVS interview.
    "PEELPAINT_LARGE", # Indicates if any section of broken plaster or peeling paint inside the unit at the time of the NYCHVS interview was at least 8 1/2 x 11 inches in size. 
    "HPROBCOUNT" # Count of select housing problems in housing unit including: 1+ heating breakdown of 6 hours or more, insufficient heat, leaks, rodents in the unit or building, toilet(s) not working, holes in walls or floors, large section(s) of peeling paint or broken plaster.
    ]
```

```python
housing_cost_columns = [
    'RENT_AMOUNT' # Monthly rent due last month, inclusive of any fees paid separately and regardless of whether this amount was paid by the tenant(s) or others on behalf of the tenant. 
]
```

```python
income_columns = [
    'HHINC_REC1' # Total household income from all persons and all sources of income in 2022.
]
```

```python
subsidy_programs_columns = [
    'RENTASSIST', # Respondent reported receiving one or more of the following types of assistance: Section 8/Housing Choice Voucher, Shelter Allowance, City FHEPS, SCRIE, DRIE, or another city, state, or federal program that pays all or part of the rent.
    'RENTASSIST_VOUCHER', # Household reported having Section 8/Housing Choice Voucher
]
```

```python
household_characteristics_columns = [
    'HHSIZE', # Indicates the number of current occupants living in the housing unit. 
    'HH62PLUS', # Indicates whether there is one or more person age 62 or older living in the housing unit.
    'HHUNDER18' # Indicates whether there is one or more person under age 18 in living in the housing unit.
]
```

```python
unit_level_survey_weight_columns = [
    'FW'
]
```

```python
replicate_weights_columns = [f"FW{i}" for i in range(1, 81)]
```

```python
occupied_dataset_subset = occupied_dataset[['CONTROL'] + housing_quality_columns + housing_cost_columns + income_columns + subsidy_programs_columns + household_characteristics_columns + unit_level_survey_weight_columns + replicate_weights_columns]
occupied_dataset_subset
```

```python
occupied_dataset_subset.to_csv("../data/processed/occupied_dataset_subset.csv", index=False)
```

# Persons dataset

```python
persons_dataset = pd.read_csv("../data/raw/person_puf_23.csv")
persons_dataset
```

```python
racial_classification_columns = [
    'RACE_P', # Indicates the person's race. 
    'RACEETH_P' # Indicates the person's race and ethnicity, combined into a single variable. 
]
```

```python
citizenship_status_columns = [
    'PLACEBORN_NYC_P', # Indicates where the person was born. 
    'PLACEBORN_MOTHER_P', # Indicates the person's mother's place of birth, by region. 
    'PLACEBORN_FATHER_P', # Indicates the person's father's place of birth, by region.
]
```

```python
marrital_status_columns = [
    "RELATION_P", # Indicates the relationship of each co-resident person in the household to the respondent.
    "MARITALSTAT_P", # Marital status of the respondent. This item is not collected for other persons in the household.
    "SPOUSE_RES_P", # Indicates where the spouse of the respondent lives. 
    "SPOUSE_P", # Indicates if the person has a spouse or partner in the household.
    "LNSPOUSE_P", # This variable indicates the line number (LNO) of co-resident spouse. 
    "LNPARTNER_P" # This variable indicates the line number (LNO) of co-resident partner. 
]
```

```python
person_level_survey_weight_columns = [
    'PW'
]
```

```python
persons_replicate_weights_columns = [f"PW{i}" for i in range(1, 81)]
```

```python
person_dataset_subset = persons_dataset[['CONTROL'] + racial_classification_columns + marrital_status_columns + person_level_survey_weight_columns + persons_replicate_weights_columns]
person_dataset_subset
```

```python
person_dataset_subset.to_csv("../data/processed/person_dataset_subset.csv", index=False)
```

# All units

```python
all_units_dataset = pd.read_csv("../data/raw/allunits_puf_23.csv")
all_units_dataset
```

```python
housing_type_columns = [
    'CSR', # Classifies the unit as one of several major types of housing based on a combination of information, including the unit's occupancy status and the current occupant(s) tenure, the regulatory status of the unit and building, and the ownership structure of the property.
    'OCC' # Occupancy status of housing unit.
]
```

```python
property_characteristics_columns = [
    'YEARBUILT', # Year in which the building where the unit is located was built. 
    'UNITS', # Indicates the total number of residential units in the building. In cases where the residential complex includes multiple buildings, this variable does not count units located in other parts of the complex.
    'ROOMS', # Number of rooms, inclusive of bedrooms, in the unit. 
]
```

```python
location_columns = [
    'BORO', # Borough (county name) where the housing unit is located 
]
```

```python
unit_level_survey_weight_columns = [
    'FW'
]
```

```python
replicate_weights_columns = [f"FW{i}" for i in range(1, 81)]
```

```python
all_units_dataset_subset = all_units_dataset[['CONTROL'] + housing_type_columns + property_characteristics_columns + location_columns + unit_level_survey_weight_columns + replicate_weights_columns]
all_units_dataset_subset
```

```python
all_units_dataset_subset.to_csv("../data/processed/all_units_dataset_subset.csv", index=False)
```
